# Figure & Ax 相關設定

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-1, 2, 10)
```

# 基本座標軸設定

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
#設定label
plt.xlabel('x')
plt.ylabel('y')

#設定座標範圍
plt.xlim(0, 2.2)
plt.ylim(-1,9)

#0~2刻度5個
plt.xticks(np.linspace(0, 2, 5))

#刻度大於範圍,範圍會變大
plt.yticks([0,2,4,6,7,8,9,10])
plt.title("Simple Plot")
plt.legend()
```

也可以透過ax方式

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
ax = plt.gca()
ax.set_xlabel('x')  # Add an x-label to the axes.
ax.set_ylabel('y')  # Add a y-label to the axes.
ax.set_title("Simple Plot")

#設定座標範圍
ax.set_xlim(0, 2.2)
ax.set_ylim(-1,9)
# equal above
#ax.set(xlim=(0, 2.2), ylim=(-1, 9))

#0~2刻度5個
ax.set_xticks(np.linspace(0, 2, 5))

#刻度大於範圍,範圍會變大
ax.set_yticks([0,2,4,6,7,8,9,10])

ax.set_title("Simple Plot")
ax.legend()
```

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')

ax = plt.gca()
ax.set(xlabel='x', ylabel='y',title='Simple Plot',
      xlim=(0, 2.2), ylim=(-1, 9),
      xticks = np.linspace(0, 2, 5),yticks =[0,2,4,6,7,8,9,10]
      )
ax.grid()
ax.legend()
```

## tick

### 設定tick方向

使用Axes.tick_params(self, axis='both', **kwargs),有４個可以用參數
<ul>
    <li>labelbottom:default True</li>
    <li>labeltop:default False</li>
    <li>labelleft:default True</li>
    <li>labelright:default False</li>
</ul>

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')
ax.tick_params(labeltop = True,labelleft = False,labelright = True)
ax.legend()
```

使用Axes.yaxis.set_tick_params()和Axes.xaxis.set_tick_params()

等於使用Axes.tick_params(self, axis='both', **kwargs)設定axix = 'x' or  axix = 'y'

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')
ax.yaxis.set_tick_params(labelleft=False, labelright=True)
ax.xaxis.set_tick_params(labeltop=True)
ax.legend()
```

使用Axes.xaxis.set_ticks_position和Axes.yaxis.set_ticks_position

<ul>
    <li>Axes.xaxis.set_ticks_position:{'top', 'bottom', 'both', 'default', 'none'}</li>
    <li>Axes.yaxis.set_ticks_position:{'left', 'right', 'both', 'default', 'none'}</li>
</ul>

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')

#將座標軸改變方向
#可是上面沒label
ax.xaxis.set_ticks_position('default')
ax.yaxis.set_ticks_position('right')
ax.legend()
```

### 設定tick格式

使用Axes.tick_params(self, axis='both', **kwargs)
<ul>
    <li>direction:{'in', 'out', 'inout'}</li>
    <li>length:Tick length in points.</li>
    <li>width:Tick width in points.</li>
    <li>colors:Tick color; accepts any mpl color spec.</li>
    <li>labelsize:Tick label font size in points or as a string (e.g., 'large').</li> 
    <li>labelrotation:Tick label rotation</li>  
</ul>


```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.tick_params(direction='in', length=3, width=1, colors='r',labelsize = '12')
ax.legend()
```

如果要分開x,y軸

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.tick_params(axis ='x',direction='in', length=3, width=1, colors='r',labelsize = '12')
ax.tick_params(axis ='y',direction='out', length=4, width=3, colors='g',labelsize = '14',labelrotation=45)
ax.legend()
```

也可使用

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.tick_params(axis ='x',direction='in', length=3, width=1, colors='r',labelsize = '12')
ax.yaxis.set_tick_params(direction='out', length=4, width=3, colors='g',labelsize = '14')
```

需要個別設定可以使用這種方法

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')
ax.legend()

for label in ax.xaxis.get_ticklabels():
    # label is a Text instance
    label.set_color('red')
    label.set_rotation(45)
    label.set_fontsize(16)

for label in ax.yaxis.get_ticklabels():
    # label is a Text instance
    label.set_color('red')
    label.set_fontsize(16) 


for line in ax.yaxis.get_ticklines():
    # line is a Line2D instance
    line.set_color('green')
    line.set_markersize(25)
    line.set_markeredgewidth(3)    
```

### 設定grid格式

使用Axes.tick_params(self, axis='both', **kwargs)
<ul>
    <li>grid_color </li>
    <li>grid_alpha </li>
    <li>grid_linewidth </li>
    <li>grid_linestyle </li> 
</ul>

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.tick_params(grid_color='r', grid_alpha=0.5) 
ax.grid()
```

x軸,y軸分開設定

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.tick_params('x',grid_color='r', grid_alpha=0.5) 
ax.tick_params('y',grid_color='g', grid_alpha=1,grid_linewidth = 1,grid_linestyle ='--') 
ax.grid()
```

一樣可以使用ax.yaxis.set_tick_params()和ax.xaxis.set_tick_params()

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.yaxis.set_tick_params(grid_color='r', grid_alpha=0.5) 
ax.xaxis.set_tick_params(grid_color='g', grid_alpha=1,grid_linewidth = 1,grid_linestyle ='--') 
ax.grid()
```

## 設定背景顏色

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')
ax.legend()

#設定fig顏色
fig.patch.set_facecolor('lightgoldenrodyellow')    
ax.patch.set_facecolor('lightslategray')
```

### position

```python
x = np.linspace(-1, 2, 10)
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')

#左邊下面框移到0位置
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.legend()
```

搭配取消邊框

```python
x = np.linspace(-1, 2, 10)
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')

#左邊下面框移到0位置
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
#取消右邊上面
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.legend()
```
