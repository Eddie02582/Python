# Figure 相關設定

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-1, 2, 10)
```

# 基本座標軸設定

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
#設定label
plt.xlabel('x')
plt.ylabel('y')

#設定座標範圍
plt.xlim((0, 2.2))
plt.ylim([-1,9])

#0~2刻度5個
plt.xticks(np.linspace(0, 2, 5))

#刻度大於範圍,範圍會變大
plt.yticks([0,2,4,6,7,8,9,10])
plt.title("Simple Plot")
plt.legend()
```

也可以透過ax方式

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
ax = plt.gca()
ax.set_xlabel('x')  # Add an x-label to the axes.
ax.set_ylabel('y')  # Add a y-label to the axes.

#設定座標範圍
ax.set_xlim((0, 2.2))
ax.set_ylim([-1,9])

#0~2刻度5個
ax.set_xticks(np.linspace(0, 2, 5))

#刻度大於範圍,範圍會變大
ax.set_yticks([0,2,4,6,7,8,9,10])

ax.set_title("Simple Plot")
ax.legend()
```

## 設定邊框顏色

這邊使用subplots()示範

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')

#取消顏色相當於取消邊框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.legend()
```

## 設定座標軸方向

```python
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')

#將座標軸改變方向
ax.xaxis.set_ticks_position('top')
ax.yaxis.set_ticks_position('right')
ax.legend()
```

## 移動座標軸

```python
x = np.linspace(-1, 2, 10)
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')

#左邊下面框移到0位置
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.legend()
```

搭配取消邊框

```python
x = np.linspace(-1, 2, 10)
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')

#左邊下面框移到0位置
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
#取消右邊上面
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.legend()
```

### 設定legend

```python
x = np.linspace(-1, 2, 10)
fig,ax = plt.subplots()
ax.plot(x, x, label='linear',linewidth=5)
ax.plot(x, x**2, label='quadratic')

'''
 'best' : 0,          
 'upper right'  : 1,
 'upper left'   : 2,
 'lower left'   : 3,
 'lower right'  : 4,
 'right'        : 5,
 'center left'  : 6,
 'center right' : 7,
 'lower center' : 8,
 'upper center' : 9,
 'center'       : 10,
'''
ax.legend(loc='upper right')
```

## 添加注释

```python
x = np.linspace(-1, 2, 10)
fig,ax = plt.subplots()
ax.plot(x, x, label='linear')

#左邊下面框移到0位置
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
#取消右邊上面
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.legend()

#plt.text(x,y,text,fontdict)
plt.text(-1, 3, r'This is the some text.',fontdict={'size': 16, 'color': 'r'})
```

```python
x = np.linspace(-1, 2, 10)
fig,ax = plt.subplots()
ax.plot(x, x, label='linear')

#左邊下面框移到0位置
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
#取消右邊上面
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.legend()

x0 = 1
y0 = x0
plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5)
# set dot styles
plt.scatter([x0, ], [y0, ], s=50, color='r')

plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))
```
