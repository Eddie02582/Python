
# Geographic Projections


This shows 4 possible geographic projections.  Cartopy_ supports more
projections.



projection{None, 'aitoff', 'hammer', 'lambert', 'mollweide', 'polar', 'rectilinear', str}

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.subplot(projection="aitoff")
plt.title("Aitoff")
plt.grid(True)
```

```python
plt.figure()
plt.subplot(projection="hammer")
plt.title("Hammer")
plt.grid(True)
```

```python
plt.figure()
plt.subplot(projection="lambert")
plt.title("Lambert")
plt.grid(True)
```

```python
plt.figure()
plt.subplot(projection="mollweide")
plt.title("Mollweide")
plt.grid(True)

plt.show()
```

```python
plt.figure()
plt.subplot(projection="polar")
plt.title("Polar")
plt.grid(True)

plt.show()
```

```python
plt.figure()
plt.subplot()
plt.title("Rectilinear")
plt.grid(True)
plt.show()

```
