## plot

主要有2種方式
<ul>
    <li>plot([x], y, [fmt], *, data=None, **kwargs)</li>
    <li>plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)</li>
</ul>

### plot([x], y, [fmt], *, data=None, **kwargs)

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

如何畫多個線

```python
x = np.linspace(0, 2, 50)
plt.plot(x, x)
plt.plot(x, x**2)
plt.plot(x, x**3)
```

### plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)

```python
x = np.linspace(0, 2, 50)
plt.plot(x, x,x, x**2,x, x**3)
```

### Other Parameter

#### label

需要搭配legend

```python
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='quadratic')
#亦可使用
#line = plt.plot(x, x**3)
#plt.setp(line, label='cubic')
plt.legend()
```

使用plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)

```python
lines = plt.plot(x, x,x, x**2,x, x**3)
plt.setp(lines[0], label='linear')
plt.setp(lines[1], label='quadratic')
plt.setp(lines[2], label='cubic')
plt.legend()
```

#### linewidth

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.legend()
```

### linestyle,marker,color 

可合併使用Formatting the style

```python
# marker symbol
plt.plot(x, x, label='linear',marker=">")
# marker from path
verts = [[-1, -1], [1, -1], [1, 1], [-1, -1]]
plt.plot(x, x**2, label='quadratic',marker=verts)

# regular asterisk marker
plt.plot(x, x**3, label='cubic',marker=(5, 2))
plt.legend()
```

## subplot

```python
plt.subplot(2,2,1)
plt.plot([0,1],[0,1],'r--',label='linear',linewidth=2)
plt.legend()

plt.subplot(2,2,2)
plt.plot([0,1],[0,2])

plt.subplot(223)
plt.plot([0,1],[0,3])

plt.subplot(224)
plt.plot([0,1],[0,4])
```

如何建立非對稱的

```python
#建立 2 * 1放入１個
plt.subplot(211)
plt.plot([0,1],[0,1])

plt.subplot(234)
plt.plot([0,1],[0,2])

plt.subplot(235)
plt.plot([0,1],[0,3])

plt.subplot(236)
plt.plot([0,1],[0,4])
```

## subplots

subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)[source]

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('Sharing x per column, y per row')
ax1.plot(x, y)
ax2.plot(x, y**2, 'orange')
ax3.plot(x, -y, 'g')
ax4.plot(x, -y**2, 'r')

for ax in fig.get_axes():
    ax.label_outer()
```
