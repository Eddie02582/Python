### Other Parameter

#### label

需要搭配legend

```python
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')

ax = plt.plot(x, x**3)
## 也可以使用plt.setp
plt.setp(ax, label='cubic')
plt.legend()
```

使用plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)需要搭配plt.setp

```python
lines = plt.plot(x, x,x, x**2,x, x**3)
plt.setp(lines[0], label='linear')
plt.setp(lines[1], label='quadratic')
plt.setp(lines[2], label='cubic')
plt.legend()
```

#### linewidth

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.legend()
```

### linestyle,marker,color 

可合併使用Formatting the style

```python
# marker symbol
plt.plot(x, x, label='linear',marker=">")
# marker from path
verts = [[-1, -1], [1, -1], [1, 1], [-1, -1]]
plt.plot(x, x**2, label='quadratic',marker=verts)

# regular asterisk marker
plt.plot(x, x**3, label='cubic',marker=(5, 2))
plt.legend()
```

## Figure 相關設定

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')

#設定label
plt.xlabel('x')
plt.ylabel('y')

#設定座標範圍
plt.xlim((0, 2.2))
plt.ylim([-1,9])

#設定座標刻度

#0~2刻度5個
plt.xticks(np.linspace(0, 2, 5))

#刻度大於範圍,範圍會變大
plt.yticks([0,2,4,6,7,8,9,10])

plt.legend()
```

設定邊框顏色

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

ax = plt.gca()
#取消顏色相當於取消邊框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
```

移動座標軸

```python
x = np.linspace(-1, 2, 10)
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

ax = plt.gca()

#取消右邊上面
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#左邊下面框移到0位置
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
```

設定座標軸方向

```python
#取消顏色相當於取消邊框
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

ax = plt.gca()
#設定座標軸方向 
ax.xaxis.set_ticks_position('top')
ax.yaxis.set_ticks_position('right')
```

設定legend

```python
plt.plot(x, x, label='linear',linewidth=5)
plt.plot(x, x**2, label='quadratic')
plt.xlabel('x')
plt.ylabel('y')

#設定legend位置
'''
 'best' : 0,          
 'upper right'  : 1,
 'upper left'   : 2,
 'lower left'   : 3,
 'lower right'  : 4,
 'right'        : 5,
 'center left'  : 6,
 'center right' : 7,
 'lower center' : 8,
 'upper center' : 9,
 'center'       : 10,
'''

plt.legend(loc='upper right')
```

```python
l1,= plt.plot(x, x, label='linear',linewidth=5)
l2,= plt.plot(x, x**2, label='quadratic')
plt.xlabel('x')
plt.ylabel('y')
#可以透過legend改label
plt.legend(handles=[l1, l2], labels=['up', 'down'],  loc='upper left')
```

添加注释

```python
x = np.linspace(-1, 2, 10)
plt.plot(x, x, label='linear',linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

ax = plt.gca()

#取消右邊上面
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#左邊下面框移到0位置
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

x0 = 1
y0 = x0
plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5)
# set dot styles
plt.scatter([x0, ], [y0, ], s=50, color='r')

plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))

# -1,3 x,y位置
plt.text(-1, 3, r'$This\ is\ the\ some\ text.$',
         fontdict={'size': 16, 'color': 'r'})        
```

## subplot

用來將多個圖畫在同一張

```python
plt.figure()
```

```python
plt.subplot(2,2,1)
plt.plot([0,1],[0,1])

plt.subplot(2,2,2)
plt.plot([0,1],[0,2])

plt.subplot(223)
plt.plot([0,1],[0,3])

plt.subplot(224)
plt.plot([0,1],[0,4])
```

```python
plt.subplot(211)
plt.plot([0,1],[0,1])

plt.subplot(234)
plt.plot([0,1],[0,2])

plt.subplot(235)
plt.plot([0,1],[0,3])

plt.subplot(236)
plt.plot([0,1],[0,4])

plt.show()  # 展示
```
