# Data Filter

```python
import pandas as pd
```

```python
path = "data.csv"
```

```python
df = pd.read_csv(path)
```

## Filter Method

<ul>
    <li>DataFrame</li>
    <li>query</li>
    <li>loc</li>
</ul>

## DataFrame

使用方法df[filter_condition]

```python
#equal filter1 = df['PTS'] > 26
filter1 = df.PTS > 26
```

```python
df[filter1]
```

如果需要多個篩選

```python
filter2 = df.EFF > 30
```

```python
df[filter1 &filter2]
```

也可以直接寫在裡面

```python
df[(df.PTS > 26) & (df['EFF'] > 30)] 
```

也可以使用lambda 表示是當作filter condition

```python
df[lambda x : x['EFF'] >30]
```

```python
df[ df.apply(lambda x : x['EFF'] >30 and x['PTS'] >26,axis = 1)]
```

也可寫成function,如果需要自訂複雜的過濾條件建議寫成function

```python
def filter_condition(x):
    return  x['EFF'] >30 and x['PTS'] >26
```

```python
df[df.apply(lambda x : filter_condition(x),axis = 1)]
```

## Query

df.query(expr),expr為string

<strong>Note:如果欄位有空白用``夾住</strong>

```python
df.query('PTS > 26 & EFF > 30')
```

```python
df.query('EFF > `PTS`+ 5')
```

## loc

跟 dataframe 使用一樣

```python
df.loc[(df.PTS > 26) & (df.EFF > 30)]
```

```python
df.loc[(df['PTS'] > 26) & (df.EFF > 30)]
```

一樣可以搭配lambda

```python
df.loc[ lambda x : x['EFF'] >30]
```

## Filter with Operator

前面都使用基本的運算式,pandas 還有提供其他的可以使用

### df[col_name].lt (lt,le,gt,ge,eq)

```python
# filter greater equal 30 
df[df["EFF"].ge(30)]
```

### df[col_name].between(a,b)

```python
df[df["GP"].between(55,60)]
```

### df[col_name].isnull()

### df[col_name].notnull()

### df[col_name].isin(array)

```python
df[df["Player"].isin(['LeBron James','Anthony Davis'])]
```

### df[col_name].str.contains(str, na=False)

```python
df[df["Player"].str.contains("James", na=False)]
```

使用 |　多個條件篩選

```python
df[df["Player"].str.contains("James|Davis|Luka", na=False)]
```
