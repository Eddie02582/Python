```python
import pandas as pd
```

```python
pd.set_option("display.max_rows",10)
```

```python
df1 = pd.read_csv("data - 1.csv")
```

```python
df2 = pd.read_csv("data - 2.csv")
```

```python
df2
```

如果資料欄位相同

```python
pd.set_option("display.max_rows",20)
```

這邊就各取前10筆結合

```python
pd.concat([df1[:10],df2[:10]])
```

可以發現index亂掉,需要使用ignore_index = True

```python
pd.concat([df1[:10],df2[:10]],ignore_index = True)
```

可以搭配keys

```python
pd.concat([df1[:5],df2[:5]], keys=['2019-2020', '2018-2019'])
```

Label the index keys you create with the names option.

```python
pd.concat([df1[:5], df2[:5]], keys=['2019-2020', '2018-2019'] , names = ["Season"])
```
