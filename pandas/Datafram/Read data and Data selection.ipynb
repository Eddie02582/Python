取得前五項



```python
df.iloc[:5]
```

```python
df.iloc[0]
```

取出前五筆和欄位前五筆

```python
df.iloc[0:5,0:5]
```

```python
df.iloc[[1,2],[0,1,2,4]]
```

```python
df.iloc[0:5,[0,1,2,4]]
```

### DataFrame.loc

取得index　= 0　到　index = 4 資料

```python
df.loc[0:4]
```

取得特定index

```python
df.loc[[0,3,4]]
```

取出index 為2的資料

```python
df.loc[2]
```

取出特定row index 和欄位

```python
df.loc[[1,3,5],['Player','GP','MIN','EFF','PTS']]
```

```python
df.loc[:2,['Player','GP','MIN','EFF','PTS']]
```

### DataFrame.loc vs DataFrame.iloc

iloc列是以當前第幾筆資料取,欄位也是

loc列是以本來的index資料取,欄位是以名稱

### 取出資料

下面可以發現iloc類似切片,loc是取出index = 0 ~ 3前面的資料

```python
df.iloc[:3]
```

```python
df.loc[:3]
```

將排序可以發現明顯的差異

```python
df_sort = df.sort_values(by='EFF',ascending=False)
```

```python
df_sort
```

iloc 取出了排序完的前3筆

```python
df_sort.iloc[:3]
```

loc取出了index 0 到　index 3的區間

```python
df_sort.loc[0:3]
```

如果順序先返loc取不到値

```python
df_sort.loc[7:4]
```
