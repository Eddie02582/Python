# Pivot table

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('example.csv')
```

```python
df
```

```python
data = pd.pivot_table(df,index='ATTENUATOR',values = ["Thruput"],columns = ['Channel','DIR'],aggfunc = np.max)
```

```python
data
```

```python
data.plot(kind="line",grid=True).set_title("ATTENUATOR vs ThroughtPut")
```

```python
pd.pivot_table(df,index=['ATTENUATOR','Channel'],values = ["Thruput"],columns = ['DIR'],aggfunc = np.max)
```
